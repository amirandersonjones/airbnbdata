```python
#Which neighorhood group has the highest average price?

air_bnb.groupby('neighbourhood_group').mean()[['price']].sort_values('price', ascending = False)
#Manhatten has the highest average price at 196.87, Bronx coming in last
```

```python
# Which neighbor hood group has the highest total price?
air_bnb.groupby(['neighbourhood_group'])[['price']].sum().sort_values(['price'], ascending=False)
#The neighborhood group with the highest total price is Manhattan @ 4264572, The lowest is Staten Island with 42825
```

```python
#Which top 5 hosts have the highest total price?



air_bnb.groupby(['host_name'])[['price']].sum().sort_values(['price'], ascending=False).head()
#The top 5 host with the highest total price is Sonder(NYC) 82,795 and Blueground is 2nd @ 70,331
```
